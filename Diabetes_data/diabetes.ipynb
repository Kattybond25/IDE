```python
import pandas as pd
diabetes = pd.read_csv('C:/Users/HONOR/OneDrive/Рабочий стол/IDE/Diabetes_data/diabetes_data.csv', sep =',')
#diabetes.head()
duplicates=diabetes[diabetes.duplicated()]
data1=diabetes.drop_duplicates()
#print(data)
threshold=0.95
informative_features=[]
for column in data1.columns:
    value_counts=data1[column].value_counts(normalize=True)
    if value_counts.max()>=threshold:
        informative_features.append(column)
data=data1[informative_features]
print(informative_features)
data_diabetes=data1.drop(informative_features, axis=1)
print(data_diabetes)
import numpy as np
columns_to_replace=['Glucose', 'BloodPressure', 'SkinThickness', 'Insulin', 'BMI']
data_diabetes[columns_to_replace]=data_diabetes[columns_to_replace].replace(0, np.nan)
missing_insulin_percentage=(data_diabetes['Insulin'].isnull().sum()/len(data_diabetes['Insulin']))
print(missing_insulin_percentage)
#Удаляем признаки где более 30% пропущенных значений
threshold=0.30
data_diabetes=data_diabetes.dropna(thresh=len(data)*(1-threshold), axis=1)
clear_data_diabetes=len(data_diabetes.columns)
print(clear_data_diabetes)
data_diabetes=data_diabetes.dropna(thresh=len(data_diabetes.columns)-2)
new_data_diabets=len(data_diabetes)
print(new_data_diabets)
data_diabetes=data_diabetes.fillna(data_diabetes.median())
mean_skin_thickness=data_diabetes['SkinThickness'].mean()
print(mean_skin_thickness)
#Выброс 1
Q1=data_diabetes['SkinThickness'].quantile(0.25)
Q3=data_diabetes['SkinThickness'].quantile(0.75)
IQR=Q3-Q1
lower_bound=Q1-1.5*IQR
upper_bound=Q3+1.5*IQR
outliers=data_diabetes[(data_diabetes['SkinThickness']<lower_bound)|(data_diabetes['SkinThickness']>upper_bound)]
outliers_count=len(outliers)
print(outliers_count)
#Выброс 2
mean_skin_thickness=data_diabetes['SkinThickness'].mean()
std_skin_thickness=data_diabetes['SkinThickness'].std()
z_threshold=3
outliers=data_diabetes[(data_diabetes['SkinThickness']-mean_skin_thickness).abs()/std_skin_thickness>z_threshold]
print(len(outliers))
# Выброс3
# Вычислите первый и третий квартили
Q1 = data_diabetes['DiabetesPedigreeFunction'].quantile(0.25)
Q3 = data_diabetes['DiabetesPedigreeFunction'].quantile(0.75)

# Вычислите межквартильный размах (IQR)
IQR = Q3 - Q1

# Определите границы для определения выбросов
lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR

# Найдите выбросы
outliers = data_diabetes[(data_diabetes['DiabetesPedigreeFunction'] < lower_bound) | (data_diabetes['DiabetesPedigreeFunction'] > upper_bound)]

# Посчитайте количество выбросов
print(len(outliers))     
# Выбросы 4
import numpy as np

# Примените логарифмическое преобразование к данным
data_diabetes['Log_DiabetesPedigreeFunction'] = np.log(data_diabetes['DiabetesPedigreeFunction'])

# Вычислите первый и третий квартили для логарифмированного признака
Q1 = data_diabetes['Log_DiabetesPedigreeFunction'].quantile(0.25)
Q3 = data_diabetes['Log_DiabetesPedigreeFunction'].quantile(0.75)

# Вычислите межквартильный размах (IQR) для логарифмированного признака
IQR = Q3 - Q1

# Определите границы для определения выбросов
lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR

# Найдите выбросы в логарифмическом масштабе
outliers = data_diabetes[(data_diabetes['Log_DiabetesPedigreeFunction'] < lower_bound) | (data_diabetes['Log_DiabetesPedigreeFunction'] > upper_bound)]

# Посчитайте количество выбросов в логарифмическом масштабе
outliers_count = len(outliers)

print(f"Количество выбросов в столбце DiabetesPedigreeFunction (в логарифмическом масштабе): {outliers_count}")           




```
